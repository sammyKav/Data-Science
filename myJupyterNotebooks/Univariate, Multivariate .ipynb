```python
import matplotlib.lines as mlines

fig, ax = plt.subplots(figsize=(10.5, 7.2)) # change the size of the image. 
SVI = data.SVI
unsure= data.unsure
plt.figure()
sns.regplot(SVI, unsure, scatter_kws={"color":"lightblue"}  ,ax = ax,line_kws={"color":"red"}) # show line of best fit. 
# shows positive relationship that suggest higher vaccine hesitancy in more vulnerable areas. 
```
